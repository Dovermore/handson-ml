```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
```

```python
def func(x,y):
    return x * y

x = np.linspace(-50, 50, 1000).reshape(-1, 1)
y = np.linspace(-50, 50, 1000).reshape(1, -1)
# z = x * y
X, Y = np.meshgrid(x.reshape(-1), y.reshape(-1))
z = np.array([func(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = z.reshape(X.shape)
```

```python
x.reshape(-1).shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="jet")
plt.show()
```

```python
def fun(x, y):
  return x**2 + y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
import tensorflow as tf
```

```python
graph1 = tf.Graph()
with graph1.as_default():
    x1 = tf.Variable(10, name='x')
    y1 = tf.Variable(5, name='y')
    op1 = x1 * y1

graph2 = tf.Graph()
with graph2.as_default():
    x2 = tf.Variable(300, name='x')
    y2 = tf.Variable(500, name='y')
    op2 = x2 * y2

session1 = tf.Session(graph=graph1)

with session1.as_default() as sess:
    print(sess)
    print(tf.get_default_session())
    with graph1.as_default():
        sess.run(tf.global_variables_initializer())
        op1.eval()
```

```python
print(tf.get_default_graph())
print(graph1)
print(graph2)

with graph2.as_default():
#     print(tf.get_default_graph())
    print(tf.get_default_graph().get_operations())
print(tf.get_default_graph())    
tf.reset_default_graph()
print(tf.get_default_graph())
with graph2.as_default():
    print(tf.get_default_graph().get_operations())
with graph1.as_default():
    print(tf.get_default_graph().get_operations())
```
